**Question 1: What is Simple Linear Regression?**
 - Simple Linear Regression is a basic and widely used statistical technique that helps us understand the relationship between two variables. In this method, one variable is taken as an independent variable and the other as a dependent variable. The main idea of simple linear regression is to find how the dependent variable changes when the independent variable changes.

In simple linear regression, the relationship between the two variables is assumed to be linear, which means it can be represented using a straight line. This straight line is known as the regression line. The regression line is fitted in such a way that it shows the best possible relationship between the variables by minimizing the difference between the actual values and the predicted values.

**Question 2: What are the key assumptions of Simple Linear Regression?**
- Simple Linear Regression works properly only when certain assumptions about the data are satisfied. These assumptions help ensure that the results obtained from the regression model are reliable and meaningful. The key assumptions of simple linear regression are explained below.

- The first assumption is linearity. This means that there should be a linear relationship between the independent variable and the dependent variable. In other words, the change in the dependent variable should be proportional to the change in the independent variable. This relationship can usually be checked using a scatter plot.

- The second assumption is independence of observations. This means that the data points should be independent of each other, and the value of one observation should not influence another. This assumption is especially important when dealing with time-based or sequential data.

- The third assumption is homoscedasticity. It means that the variance of the error terms remains constant for all values of the independent variable. In simple words, the spread of residuals should be roughly the same across the regression line. If this condition is not met, it can affect the accuracy of predictions.

- The fourth assumption is normality of errors. This assumes that the residuals (differences between actual and predicted values) are normally distributed. Normality is important for making valid statistical inferences such as confidence intervals and hypothesis testing.

- The final assumption is no significant outliers. Extreme values can have a strong influence on the regression line and may lead to misleading results. Therefore, the dataset should be checked for outliers before applying simple linear regression.

- When these assumptions are reasonably satisfied, simple linear regression provides accurate predictions and meaningful insights into the relationship between variables.

**Question 3: What is heteroscedasticity, and why is it important to address in regression models?**
- Heteroscedasticity refers to a situation in regression analysis where the variance of the error terms (residuals) is not constant across all values of the independent variable. In simple terms, it means that the spread of the residuals changes as the value of the independent variable changes. This violates one of the key assumptions of simple linear regression, which assumes constant variance, also known as homoscedasticity.

- When heteroscedasticity is present, the residuals may show a funnel-shaped or uneven pattern when plotted against the independent variable or predicted values. For example, in a model predicting income based on work experience, the variation in income might be small for beginners but much larger for highly experienced individuals, leading to unequal variance of errors.

- Addressing heteroscedasticity is important because it affects the reliability of the regression results. Although the regression coefficients may still be unbiased, the standard errors become inaccurate. This can lead to incorrect hypothesis testing, misleading confidence intervals, and wrong conclusions about the significance of variables.

- If heteroscedasticity is ignored, the model may appear to perform well, but the predictions and statistical interpretations may not be trustworthy. Therefore, it is important to detect and correct it using methods such as data transformation (like log transformation), using weighted least squares, or applying robust standard errors.

The main difference between linear regression and polynomial regression lies in the nature of the relationship they model. Linear regression assumes a straight-line relationship between variables, making it suitable when data follows a simple trend. Polynomial regression, on the other hand, is used when data shows a curved pattern that linear regression cannot accurately capture.

Another important difference is flexibility. Polynomial regression is more flexible and can fit complex data patterns, but this also increases the risk of overfitting if the degree of the polynomial is too high. Linear regression is simpler, easier to interpret, and less likely to overfit.

In practical applications, polynomial regression is often used in scenarios such as modeling growth trends, predicting changes in temperature over time, or analyzing non-linear business or scientific data. Overall, polynomial regression provides a powerful alternative when linear regression is not sufficient to explain the relationship between variables.


```python
#Question 6: Implement a Python program to fit a Simple Linear Regression model to the following sample data:
# X = [1, 2, 3, 4, 5] Y = [2.1, 4.3, 6.1, 7.9, 10.2]
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and train the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Plot data points and regression line
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Simple Linear Regression")
plt.show()

# Print model parameters
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
#Question 7: Fit a Multiple Linear Regression model on this sample data:
# Area = [1200, 1500, 1800, 2000] ‚óè Rooms = [2, 3, 3, 4] ‚óè Price = [250000, 300000, 320000, 370000]
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Given data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Independent and dependent variables
X = df[['Area', 'Rooms']]
Y = df['Price']

# Add constant for VIF calculation
X_const = sm.add_constant(X)

# Fit Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Model coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i)
                   for i in range(X_const.shape[1])]

print("\nVIF Values:")
print(vif_data)

```

```python
#Question 8: Implement polynomial regression on the following data:
# X = [1, 2, 3, 4, 5] ‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Generate predictions
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
Y_pred = model.predict(X_range_poly)

# Plot original data and polynomial curve
plt.scatter(X, Y)
plt.plot(X_range, Y_pred)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("2nd Degree Polynomial Regression")
plt.show()

# Print model coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

```

```python
#Question 9: Create a residuals plot for a regression model trained on this data:
# X = [10, 20, 30, 40, 50] ‚óè Y = [15, 35, 40, 50, 65]
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train regression model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals
plt.scatter(X, residuals)
plt.axhline(y=0)
plt.xlabel("X values")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()

# Print residuals
print("Residuals:", residuals)

```

Question 10: Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location.However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model?
- As a data scientist working for a real estate company, building a reliable house price prediction model is very important because business decisions depend on it. If heteroscedasticity and multicollinearity are detected in the regression model, specific steps must be taken to correct these issues and make the model more robust and trustworthy.

To address heteroscedasticity, the first step would be to visually confirm it using residual plots. If the spread of residuals increases or decreases with predicted values, it indicates non-constant variance. One common solution is to apply data transformations such as log transformation on the target variable (house price) or on highly skewed features like area. This often stabilizes variance and improves model performance. Another effective approach is to use robust standard errors or weighted least squares regression, which reduce the impact of unequal error variance and lead to more reliable statistical inference.

For handling multicollinearity, the first step would be to calculate correlation matrices and Variance Inflation Factor (VIF) values to identify highly correlated independent variables. If two or more variables such as area and number of rooms show high VIF values, one of them can be removed or combined based on domain knowledge. Feature selection techniques like backward elimination or regularization methods such as Ridge Regression and Lasso Regression are also useful, as they reduce the effect of correlated variables by shrinking coefficients.

Additionally, proper feature engineering plays a key role. For example, creating price-per-square-foot instead of using raw area and rooms separately may reduce redundancy. Standardizing or normalizing features can also help improve numerical stability of the model.

Finally, after applying these corrections, the model should be re-evaluated using performance metrics such as R¬≤, Adjusted R¬≤, and RMSE, along with cross-validation to ensure consistent performance on unseen data. By systematically addressing heteroscedasticity and multicollinearity, the regression model becomes more stable, interpretable, and suitable for real-world house price prediction.
